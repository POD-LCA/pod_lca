```python
class Project_logestic_manager:


    def __init__(self, name, location, data_folder):

        self.name = name
        self.link = []
        self.impact = {'GWP':0.0, 'AP':0.0, 'EP':0.0, 'ODP':0.0, 'SFP':0.0}
        self.location = location
        self.data_folder= data_folder
        self.subdataset = {}

    def sub_dataset(self):
        
        for file in os.listdir(self.data_folder):
            file_path = os.path.join(self.data_folder, file)
            
            if file.endswith('.csv'):
                try:
                    df = pd.read_csv(file_path)
                    dataset_name = os.path.splitext(file)[0]
                    self.subdataset[dataset_name] = df
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")
            elif file.endswith(('.xlsx', '.xls')):
                try:
                    df = pd.read_excel(file_path)
                    dataset_name = os.path.splitext(file)[0]
                    self.subdataset[dataset_name] = df
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")


    def get_subdataset(self, sub_dataset = str):

        Project_logestic_manager.sub_dataset(self)
        return self.subdataset [sub_dataset]

    def get_data_folder(self):

        return self.data_folder

    def get_impact(self, impact_cat):

        return self.impact

            
    def create_links(self, material, qty, travel_dist, return_travel_dist, disy_unit):
        
        link = Logisticslink (material, qty, travel_dist, return_travel_dist, disy_unit)

        self.link.append (link)
        self.impacts + link.compute_impact()

    def create_links_scenario (self, material, scenario):
        
        if scenario == 'Local':

            link = National ()
            self.link.append (link)
            self.impacts + link.compute_impact()

        elif scenario == 'Regional':
            link = National(scenario)

            link = Regional ()
            self.link.append (link)
            self.impacts + link.compute_impact()

        elif scenario == 'Reginal_c':
            link = Reginal_c(scenario)

            link = Reginal_c ()
            self.link.append (link)
            self.impacts + link.compute_impact()

        elif scenario == 'National':
            link = National(scenario)

            link = National ()
            self.link.append (link)
            self.impacts + link.compute_impact()

        else:
            print (f"your scenario is not defined")

    def get_link (self):
        
        return self.link

    def get_link_material (self):

        return [link.material for link in self.link]
```
